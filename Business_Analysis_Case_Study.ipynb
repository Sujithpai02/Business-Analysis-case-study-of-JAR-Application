# disparities and Improvements

<h3>1.Sikkim Observation:</h3> Very low order volume across all categories.

Likely Reason: Small population, limited logistics access.

Recommendation:

Run awareness campaigns.

Offer bundled shipping for remote areas.

Partner with local sellers to build trust.

<h3>2.Goa</h3>
Observation: Moderate Clothing orders, very low in Electronics and Furniture.

Likely Reason: Tourist-heavy state, small housing sizes.

Recommendation:

Focus on portable/foldable furniture and travel accessories.

Promote seasonal electronics (e.g. headphones, cameras).

Geo-target ads around major cities (e.g., Panaji, Vasco).

<h3>3.Punjab</h3>
Observation: Strong Clothing sales, weak in Furniture & Electronics.

Likely Reason: Preference for traditional or offline electronics/furniture.

Recommendation:

Push fashion + electronics combos (e.g. offers with wearables).

Educate users via reels/blogs on furniture quality and return policy.

<h3>4.Tamil Nadu</h3>
Observation: Decent performance in Clothing, lagging in Furniture and Electronics.

Likely Reason: Competitive regional e-commerce platforms.

Recommendation:

Collaborate with local influencers in Tamil.

Highlight product variety & delivery speed in ads.

Add regional festival-based campaigns (e.g. Pongal offers).

<h3>5. Himachal Pradesh</h3>
Observation: Very low across all categories.

Likely Reason: Hilly terrain  delivery/logistics challenges.

Recommendation:

Target urban hubs like Shimla, Dharamshala first.

Promote durable, lightweight products.

Use local delivery partners to ensure reliability.
